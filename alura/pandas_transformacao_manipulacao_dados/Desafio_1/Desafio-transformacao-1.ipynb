```python
import pandas as pd
```

```python
data = pd.read_json('dados_vendas_clientes.json')
data
```

```python
data = pd.json_normalize(data['dados_vendas'])
data
```

```python
columns = list(data.columns)[1:]
columns
```

```python
data = data.explode(columns)
data
```

```python
data.reset_index(drop=True, inplace=True)
data
```

```python
data.info()
```

```python
import numpy as np

data['Valor da compra'] = data['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip())
data['Valor da compra'] = data['Valor da compra'].astype(np.float64)
data.info()
```

```python
data['Cliente']
```

```python
data['Cliente'] = data['Cliente'].str.replace('[^a-zA-Z\u00C0-\u00FF| ]', '', regex=True)
data.head()
```

```python
data['Cliente'] = data['Cliente'].str.lower()
data.head()
```

```python
data['Data de venda'] = pd.to_datetime(data['Data de venda'], format='%d/%m/%Y')
data.info()
```

```python
df_client_value = data.groupby(data['Cliente'])[['Valor da compra']].max()
df_client_value
```
