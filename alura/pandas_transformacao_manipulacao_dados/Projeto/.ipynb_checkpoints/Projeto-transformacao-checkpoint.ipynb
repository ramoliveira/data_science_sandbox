```python
import pandas as pd
```

```python
data = pd.read_json('dados_hospedagem.json')
data.head()
```

```python
data = pd.json_normalize(data['info_moveis'])
data.head()
```

```python
columns = list(data.columns)
columns
```

```python
data = data.explode(columns[3:])
data.head()
```

```python
data.reset_index(inplace=True, drop=True)
data
```

```python
data.info()
```

```python
import numpy as np
```

```python
data['max_hospedes'] = data['max_hospedes'].astype(np.int64)
```

```python
data.info()
```

```python
numbered_columns = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
```

```python
data[numbered_columns] = data[numbered_columns].astype(np.int64)
```

```python
data.info()
```

```python
data['avaliacao_geral'] = data['avaliacao_geral'].astype(np.float64)
```

```python
data.info()
```

```python
data['preco'] = data['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
data['preco'] = data['preco'].astype(np.float64)
```

```python
data.info()
```

```python
data[['taxa_deposito', 'taxa_limpeza']] = data[['taxa_deposito', 'taxa_limpeza']].map(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
data[['taxa_deposito', 'taxa_limpeza']] = data[['taxa_deposito', 'taxa_limpeza']].astype(np.float64)
```

```python
data.info()
```

```python
data['descricao_local'] = data['descricao_local'].str.lower()
```

```python
data.head()
```

```python
data['descricao_local'] = data['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
data['descricao_local']
```

```python
data['descricao_local'] = data['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
data['descricao_local']
```

```python
data['descricao_local'] = data['descricao_local'].str.split()
data.head()
```

```python
data['comodidades'] = data['comodidades'].str.replace('\"|{|}', '', regex=True)
data['comodidades']
```

```python
data['comodidades'] = data['comodidades'].str.split(',')
data.head()
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.lower()
data['descricao_vizinhanca']
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
data['descricao_vizinhanca']
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
data['descricao_vizinhanca']
```

```python
data['descricao_vizinhanca'] = data['descricao_vizinhanca'].str.split()
data.head()
```
