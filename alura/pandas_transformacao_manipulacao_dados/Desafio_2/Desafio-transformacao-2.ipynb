```python
import pandas as pd
```

```python
data = pd.read_json('dados_locacao_imoveis.json')
data
```

```python
data = pd.json_normalize(data['dados_locacao'])
data
```

```python
columns = list(data.columns)[1:]
columns
```

```python
data = data.explode(columns)
data.head()
```

```python
data.reset_index(drop=True, inplace=True)
data
```

```python
data.info()
```

```python
import numpy as np

data['valor_aluguel'] = data['valor_aluguel'].apply(lambda x: x.replace('$', '').replace('reais', '').replace(',', '.').strip())
data['valor_aluguel'] = data['valor_aluguel'].astype(np.float64)
data.info()
```

```python
data['apartamento'] = data['apartamento'].apply(lambda x: x.replace('(blocoAP)', '').strip())
data.head()
```

```python
data['datas_combinadas_pagamento'] = pd.to_datetime(data['datas_combinadas_pagamento'], format='%d/%m/%Y')
data.info()
```

```python
data['datas_de_pagamento'] = pd.to_datetime(data['datas_de_pagamento'], format='%d/%m/%Y')
data.info()
```

```python
data['atraso'] = (data['datas_de_pagamento'] - data['datas_combinadas_pagamento']).dt.days
data['atraso']
```

```python
mean_late_payment = data.groupby(['apartamento'])['atraso'].mean()
mean_late_payment
```
