```python
df = data.fillna(value=0)
df
```

## Remoção dos alunos Alice e Carlos

```python
selection1 = df['Nome'] == 'Ana'
df[selection1]
```

```python
selection2 = df['Nome'] == 'Carlos'
df[selection2]
```

```python
selection = selection1 | selection2
df[selection]
```

```python
indexes_to_remove = df[selection].index
```

```python
df = df.drop(indexes_to_remove, axis=0)
df
```

## Aplicação de filtro

### Apenas alunos aprovados

```python
df_1 = df[(df['Aprovado'] == True)]
df_1
```

# Desafio 3

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
data = pd.read_csv(url)
data.fillna(value=0, inplace=True)
data
```

```python
data['Pontos_extras'] = data['Notas'].apply(lambda x: x * 0.40)
data.head()
```

```python
data['Notas_finais'] = data['Notas'] + data['Pontos_extras']
data.head()
```

```python
data['Aprovado_final'] = data['Notas_finais'].apply(lambda x: True if x >= 6 else False)
data.head()
```

```python
data[(data['Aprovado'] == False) & (data['Aprovado_final'])]
```
